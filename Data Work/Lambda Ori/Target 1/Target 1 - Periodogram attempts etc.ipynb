```python
from lightkurve import TessTargetPixelFile
```

```python
tpf = TessTargetPixelFile('https://mast.stsci.edu/tesscut/api/v0.1/astrocut?ra=82.78252411&dec=11.42396927&y=5&x=5&units=px&sector=All')
```

```python
tpf.plot(frame=0)
```

```python
tpf.plot(aperture_mask=tpf.pipeline_mask)
```

```python
#This code is used to create a new aperture. Please note that this code is from Lightkurve Tutorials. 
import numpy as np
# Create a median image of the source over time
median_image = np.nanmedian(tpf.flux, axis=0)
# Select pixels which are brighter than the 86th percentile of the median image
aperture_mask = median_image > np.nanpercentile(median_image, 96)
# Plot that aperture
tpf.plot(aperture_mask=aperture_mask);
```

```python
lc = tpf.to_lightcurve(aperture_mask=aperture_mask)
```

```python
lc.scatter()
```

```python
lc.remove_nans().remove_outliers(sigma=5).scatter()
```

```python
pg = lc.to_periodogram(oversample_factor=1)
```

```python
pg.plot();
```

```python
pg.plot(view='period', scale='log')
```

```python
period=pg.period_at_max_power
print("Best Period: {}".format(period))
```

```python
lc.remove_nans().remove_outliers(sigma=4).fold(period.value).flatten().scatter()
```

```python
lc.remove_nans().remove_outliers(sigma=4).fold(period.value).scatter()
```
