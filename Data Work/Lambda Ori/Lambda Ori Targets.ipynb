```python
tpf5=TessTargetPixelFile('https://mast.stsci.edu/tesscut/api/v0.1/astrocut?ra=86.90058899&dec=9.28265381&y=13&x=13&units=px&sector=All')
```

```python
tpf5.plot()
```

```python
#This code is used to create a new aperture. Please note that this code is from Lightkurve Tutorials. 
import numpy as np
# Create a median image of the source over time
median_image = np.nanmedian(tpf5.flux, axis=0)
# Select pixels which are brighter than the 86th percentile of the median image
aperture_mask = median_image > np.nanpercentile(median_image, 92)
# Plot that aperture
tpf5.plot(aperture_mask=aperture_mask);
```

```python
lc5=tpf5.to_lightcurve(aperture_mask=aperture_mask)
```

```python
lc5.scatter();
```

```python
lc5.remove_nans().remove_outliers(sigma=5).scatter();
```
