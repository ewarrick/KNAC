```python
from lightkurve import search_targetpixelfile
```

```python
tpf=search_targetpixelfile(436158814, sector=6).download()
```

```python
tpf.plot(frame=0)
```

```python
tpf.plot(aperture_mask=tpf.pipeline_mask)
```

```python
#This code is used to create a new aperture. Please note that this code is from Lightkurve Tutorials. 
import numpy as np
# Create a median image of the source over time
median_image = np.nanmedian(tpf.flux, axis=0)
# Select pixels which are brighter than the 86th percentile of the median image
aperture_mask = median_image > np.nanpercentile(median_image, 95)
# Plot that aperture
tpf.plot(aperture_mask=aperture_mask);
```

```python
lc=tpf.to_lightcurve(aperture_mask=aperture_mask)
lc.remove_outliers(sigma=5).remove_nans().bin(1).scatter();
```

```python
tpf.get_coordinates(0)
```

```python
pg = lc.to_periodogram(oversample_factor=10)
```

```python
pg.plot();
```

```python
pg.plot(view='period', scale='log')
```

```python
period = pg.period_at_max_power
print("Best Period: {}".format(period))
lc.remove_nans().remove_outliers(sigma=5).fold(period.value).scatter();
```

```python
print(lc.time.shape)
```

```python
lc.remove_nans().remove_outliers(sigma=5).flatten(window_length=1001).fold(period.value).bin(binsize=1).scatter();
```
